1.2 **Offline vs on-the-fly обработка**

> В этом задании нам предстоит превратить сырые строковые последовательности ДНК в числовое представление, пригодное для подачи в нейросеть. Существует два основных подхода к предобработке данных:
> * Offline-предобработка (всё заранее в __init__)
>   * **Идея**: при инициализации `Dataset` мы полностью подготавливаем данные и храним их уже в виде тензоров фиксированной формы.
>   * **Что делаем в `__init__`:**
>      * очищаем последовательности (оставляем только A/C/G/T/N, приводим к одному регистру);
>      * обрезаем/паддим каждую последовательность до фиксированной длины $L$ (например, 1000 нуклеотидов);
>      * кодируем каждую последовательность в One-Hot-тензор формы $L \times 4$;
>      * сохраняем полученные тензоры (например, в списке или в `torch.Tensor`).
>   * **Плюсы:**
>      * простая логика в `__getitem__` — просто возвращаем готовый тензор по индексу;
>      * нет вычислительных затрат на предобработку во время обучения;
>      * удобно для небольших по размеру задач.
>   * **Минусы:**
>      * если датасет очень большой (миллионы последовательностей), полный One-Hot всех данных может не влезть в память;
>      * любые изменения в схеме кодирования требуют пересоздания всего `Dataset`.
> * On-the-fly обработка (в __getitem__)
>   * **Идея**: в конструкторе сохраняем исходные строки или минимально очищенные данные, а кодирование в One-Hot выполняем при каждом обращении `__getitem__`.
>   * **Что делаем в `__init__`:**
>      * сохраняем список строк ДНК и таргетов;
>      * при желании — выполняем лёгкую очистку (например, фильтрацию символов).
>   * **Что делаем в `__getitem__`:**
>      * берем сырую строку по индексу;
>      * приводим к нужной длине $L$ (обрезка/паддинг);
>      * превращаем в One-Hot-тензор $L \times 4$ «на лету».
>   * **Плюсы:**
>      * не нужно хранить весь One-Hot в памяти;
>      * легко изменять схему кодирования без пересборки всего набора.
>   * **Минусы:**
>      * часть вычислений выполняется на каждой итерации обучения;
>      * требует аккуратной реализации в `__getitem__`.

1.3 **Параллельная обработка и гибрид**

1.2 **Когда графовые нейросети оправданы в задачах по ДНК**

> Графовые нейросети (GNN) особенно полезны, когда данные естественно описываются в виде графа, а не простой последовательности. Для ДНК это может быть:
> * **Граф регуляторных взаимодействий:**
>   * вершины — регуляторные элементы (энхансеры, промоторы, сайленсеры);
>   * рёбра — известные взаимодействия (enhancer–promoter, enhancer–gene и т.п.);
>   * задача — предсказывать активность или эффект мутаций с учётом сети регуляции.
> * **Граф 3D-контактов (Hi-C, TAD-домены):**
>   * вершины — участки генома (окна фиксированной длины);
>   * рёбра — пары участков с высокой частотой контактов в 3D (по данным Hi‑C);
>   * задача — моделировать влияние удалённых регуляторных регионов на таргетный ген.
> * **Граф k-меров / мотивов:**
>   * вершины — k-меры или обнаруженные мотивы;
>   * рёбра — отношения ко-встречаемости, соседства или функциональной связи между мотивами.
> Во всех этих случаях GNN помогает агрегировать информацию от соседей в графе (по рёбрам) и захватывать сложные зависимости, которые нельзя выразить только локальной «линейной» свёрткой по последовательности.
> **Важно:** чтобы построить такой граф, нужны дополнительные биологические знания и/или внешние данные (Hi‑C, ChIP-seq, базы регуляторных взаимодействий и т.д.). В рамках нашего ДЗ мы работаем только с окном последовательности и бинарным таргетом, без явной 3D-информации, поэтому GNN здесь носит скорее учебно-исследовательский характер.

4. **Ранжирование k-меров:**
> * Пусть $m \in \{\text{xai}_1, \text{xai}_2, \text{xai}_3\}$ — индекс метода интерпретации.
> * Для каждого метода интерпретации $m \in \{ \text{xai}_1, \text{xai}_2, \text{xai}_3 \}$:

>     $$
     \mu_m(\text{kmer}) = \frac{1}{N} \sum_{n=1}^{N} I_{\text{kmer},n}^{(m)}
     $$
>     где:
>     - \( N \) - количество тестовых True Positive объектов
>     - \( $I_{\text{kmer},n}^{(m)} $\) - важность k-мера в объекте \( n \) согласно методу \( m \)

>   $$
   \text{PMD}_m(\text{kmer}) = \frac{1}{N} \sum_{n=1}^{N} \frac{\left| I_{\text{kmer},n}^{(m)} - \mu_m(\text{kmer}) \right|}{\left|\mu_m(\text{kmer})\right| + \varepsilon} \times 100\%,

>   где $\varepsilon$ — небольшое число, предотвращающее деление на ноль.

> * Вы получите по три оценки (от каждого метода интерпретации) на каждый уникальный k-мер.

>   - высокая $\text{PMD}_m$ ⇒ вклад k-мера нестабилен от объекта к объекту (возможно, шум или контекст-зависимый мотив).
